```python
import requests
import pandas as pd
from pyspark.sql.functions import col, explode, from_json, to_json

# Start a new replay here: https://sportsdata.io/members/replays
api_url = "https://replay.sportsdata.io/api/v3/cbb/stats/json/boxscoresdelta/2023-12-02/all?key=43e4b6ccaeed4000bc63bc67e0c7b300"

# Define function to fetch data from the API
def fetch_data_from_api(api_url):
    response = requests.get(api_url)
    response.raise_for_status()  # Raise an error for bad status codes
    return response.json()

# Define a function to fill NaN values
# Note: Fill 0 for numeric columns, fill blank string for string columns
def fill_nas(df):
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            df[column] = df[column].fillna(0)
        elif pd.api.types.is_string_dtype(df[column]):
            df[column] = df[column].fillna('')
    return df

# Load the JSON data from the API and convert to a pandas DataFrame
json_data = fetch_data_from_api(api_url)
pdf = pd.DataFrame(json_data)

# Add a timestamp (UTC) column to the DataFrame
current_timestamp = pd.Timestamp.now()

# Explode to build the df for player_games
pdf_exploded = pdf.explode('PlayerGames')
df_player_games = pd.json_normalize(pdf_exploded['PlayerGames'])

# DQC: If player_games is empty, no need to continue the script.
# Note: API returns empty list if no games have started for a given day.
# TODO: This is only a good solution if I can intergrate with orchestration flow
# Assumes code is running via Databricks notebook; not a good solution.
if df_player_games.empty:
    dbutils.notebook.exit("Stopping processing; df_player_games is empty.")

df_player_games['timestamp'] = current_timestamp
df_player_games = fill_nas(df_player_games)
df_player_games = df_player_games[df_player_games['PlayerID'] != 0]
df_player_games = df_player_games.astype(schema_mapping_player_games)

# Normalize to build the df for games
df_games = pd.json_normalize(pdf['Game'])
df_games['timestamp'] = current_timestamp
# There is an entire nested section in the raw json devoted to stadium info, drop it
# TODO: Need to keep stadium info if I ever want lat/long data
columns_to_drop = [col for col in df_games.columns if col.startswith('Stadium')]
# Periods is another nested section that needs to be dropped
columns_to_drop.extend(['Periods'])
df_games = df_games.drop(columns=columns_to_drop)

# Fill null values
df_games = fill_nas(df_games)

# Convert to the correct data types
df_games = df_games.astype(schema_mapping_games)

# Create the Spark dataframes
df_player_games = spark.createDataFrame(df_player_games)
df_games = spark.createDataFrame(df_games)

# Create/append to Delta Lake table
# Note: Attempts to append data using @dlt.table were unsuccessful (each run overwrote the table)
df_player_games.write.format("delta").mode("append").partitionBy("timestamp").saveAsTable("tabular.dataexpert.jw_raw_player_games")
df_games.write.format("delta").mode("append").partitionBy("timestamp").saveAsTable("tabular.dataexpert.jw_raw_games")
```
